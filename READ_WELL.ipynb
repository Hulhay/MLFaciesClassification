# Jumlah Sumur yang digunakan

```python
wells = dfTrain['WELL'].unique()
print('jumlah sumur', len(wells))
wells
```

Dilakukan filter dari data latih (dfTrain) sehingga hanya menampilkan data-data dengan lokasi sumur yang sama (sumur 15/9-13).
Hal ini dilakukan untuk kebutuhan visualisasi data.

```python
wellData = dfTrain[dfTrain['WELL'] == wells[0]]
wellData
```

```python
wellData = wellData[['DEPTH_MD', 'CALI', 'RSHA', 'RMED', 'RDEP', 'RHOB', 'GR', 'SGR', 'NPHI', 'PEF', 'DTC', 'SP', 'BS', 'ROP', 'DTS', 'DCAL', 'DRHO', 'MUDWEIGHT', 'RMIC', 'ROPA', 'RXO', 'FORCE_2020_LITHOFACIES_LITHOLOGY']]
wellData.head()
```

# Plot Sumur 15/9-13

```python
wellDataVal = wellData.values
```

```python
import matplotlib.pyplot as plt
```

```python
facies = wellData['FORCE_2020_LITHOFACIES_LITHOLOGY'].values
facies
```

```python
import numpy as np
```

```python
rows, cols = 1, 21
fig, ax = plt.subplots(nrows=rows,
                      ncols=cols,
                      figsize=(22,16),
                      sharey=True
                     )

titles = 'CALI', 'RSHA', 'RMED', 'RDEP', 'RHOB', 'GR', 'SGR', 'NPHI', 'PEF', 'DTC', 'SP', 'BS', 'ROP', 'DTS', 'DCAL', 'DRHO', 'MUDWEIGHT', 'RMIC', 'ROPA', 'RXO', 'FACIES'
for i in range(cols):
    if i < cols-1:
        ax[i].plot(wellDataVal[:,i+1], wellDataVal[:,0], linewidth='0.5')
        ax[i].set_ylim(max(wellDataVal[:,0]), min(wellDataVal[:,0]))
        ax[i].minorticks_on()
        ax[i].set_title('%s' %(titles[i]))
        ax[i].grid(which='major',
                  linestyle='-',
                  linewidth='0.5',
                  color='red'
                  )
        ax[i].grid(which='minor',
                  linestyle=':',
                  linewidth='0.5',
                  color='black'
                  )
    elif i == cols-1:
        F = np.vstack((facies, facies)).T
        ax[i].imshow(F, aspect='auto', extent=[0, 1, max(wellDataVal[:,0]), min(wellDataVal[:,0])])
        ax[i].set_title('%s' %(titles[i]))

fig.suptitle('SUMUR 15/9-13', fontsize=20)
plt.show()
```
