```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fn = 'F02-1_logs.las'
dataLoad = np.loadtxt(fn, skiprows=35)
dataLoad
```

```python
dataLoad[dataLoad==-999.2500] = np.nan
dataLoad
```

```python
mnemonics = ['DEPTH', 'RHOB', 'DT', 'GR', 'AI', 'AI_rel', 'PHIE']
df = pd.DataFrame(dataLoad, columns=mnemonics)
df.head()
```

```python
df = df[['DEPTH', 'RHOB', 'DT', 'GR', 'PHIE']]
df.head()
```

```python
dataLog = df.values
```

```python
rows, cols = 1, 4
fig, ax = plt.subplots(nrows=rows,
                      ncols=cols,
                      figsize=(12,10),
                      sharey=True
                     )
titles = 'RHOB', 'DT', 'GR', 'PHIE'
for i in range(cols):
    ax[i].plot(dataLog[:,i+1], dataLog[:,0])
    ax[i].set_ylim(max(dataLog[:,0]), min(dataLog[:,0]))
    ax[i].minorticks_on()
    ax[i].set_title('%s' %(titles[i]))
    ax[i].grid(which='major',
              linestyle='-',
              linewidth='0.5',
              color='red'
              )
    ax[i].grid(which='minor',
              linestyle=':',
              linewidth='0.5',
              color='black'
              )

plt.show()
```

```python
rows, cols = 1, 3
fig, ax = plt.subplots(nrows=rows,
                      ncols=cols,
                      figsize=(12,10),
                      sharey=True
                     )
titles = 'RHOB', 'DT', 'GR'
for i in range(cols):
    ax[i].plot(dataLog[:,i+1], dataLog[:,0], linewidth='0.5')
    ax[i].set_ylim(max(dataLog[:,0]), min(dataLog[:,0]))
    ax[i].minorticks_on()
    ax[i].set_title('%s' %(titles[i]))
    ax[i].grid(which='major',
              linestyle='-',
              linewidth='0.5',
              color='red'
              )
    ax[i].grid(which='minor',
              linestyle=':',
              linewidth='0.5',
              color='black'
              )

y2 = dataLog[:,3]    
y1 = y2*0+50
ax[2].fill_betweenx(dataLog[:,0],
                    y1, 
                    y2, 
                    where=(y1>=y2), 
                    color='gold', 
                    linewidth=0
                   )
ax[2].fill_betweenx(dataLog[:,0],
                    y1, 
                    y2, 
                    where=(y1<y2), 
                    color='lime', 
                    linewidth=0
                   )
    
plt.subplots_adjust(wspace=0)
plt.show()
```

```python
tb = [0, 464, 539, 612, 635, 687, 702, 795, 814, 926, 949, 1026, 1053, 1095, 1133, 1270, 1297, 1430, 2000]
f = [1, 2, 3, 1, 3, 4, 3, 1, 3, 1, 3, 1, 3, 1, 3, 4, 3, 1]

print(len(tb))
print(len(f))
```

```python
depth = df.DEPTH.values
```

```python
facies = []

for i in range(len(depth)):
    for j in range(len(tb)-1):
        if depth[i] > tb[j] and depth[i] <= tb[j+1]:
            facies.append(f[j])

df['FACIES'] = facies
df.head()
```

```python
dfNew = df[['DEPTH', 'RHOB', 'DT', 'GR', 'FACIES']]
dfNew.head()
```

```python
dfNewValues = dfNew.values
```

```python
rows, cols = 1, 4
fig, ax = plt.subplots(nrows=rows,
                      ncols=cols,
                      figsize=(12,10),
                      sharey=True
                     )
titles = 'RHOB', 'DT', 'GR', 'FACIES'
for i in range(cols):
    ax[i].plot(dfNewValues[:,i+1], dfNewValues[:,0], linewidth='0.5')
    ax[i].set_ylim(max(dfNewValues[:,0]), min(dfNewValues[:,0]))
    ax[i].minorticks_on()
    ax[i].set_title('%s' %(titles[i]))
    ax[i].grid(which='major',
              linestyle='-',
              linewidth='0.5',
              color='red'
              )
    ax[i].grid(which='minor',
              linestyle=':',
              linewidth='0.5',
              color='black'
              )

y2 = dfNewValues[:,3]    
y1 = y2*0+50
ax[2].fill_betweenx(dfNewValues[:,0],
                    y1, 
                    y2, 
                    where=(y1>=y2), 
                    color='gold', 
                    linewidth=0
                   )
ax[2].fill_betweenx(dfNewValues[:,0],
                    y1, 
                    y2, 
                    where=(y1<y2), 
                    color='lime', 
                    linewidth=0
                   )
    
plt.subplots_adjust(wspace=0)
plt.show()
```

```python
rows, cols = 1, 4
fig, ax = plt.subplots(nrows=rows,
                      ncols=cols,
                      figsize=(12,10),
                      sharey=True
                     )
titles = 'RHOB', 'DT', 'GR', 'FACIES'
for i in range(cols):
    if i < cols-1:
        ax[i].plot(dfNewValues[:,i+1], dfNewValues[:,0], linewidth='0.5')
        ax[i].set_ylim(max(dfNewValues[:,0]), min(dfNewValues[:,0]))
        ax[i].minorticks_on()
        ax[i].set_title('%s' %(titles[i]))
        ax[i].grid(which='major',
                  linestyle='-',
                  linewidth='0.5',
                  color='red'
                  )
        ax[i].grid(which='minor',
                  linestyle=':',
                  linewidth='0.5',
                  color='black'
                  )
    elif i == cols-1:
        F = np.vstack((facies, facies)).T
        ax[i].imshow(F, aspect='auto', extent=[0, 1, max(dfNewValues[:,0]), min(dfNewValues[:,0])])
        ax[i].set_title('%s' %(titles[i]))

y2 = dfNewValues[:,3]    
y1 = y2*0+50
ax[2].fill_betweenx(dfNewValues[:,0],
                    y1, 
                    y2, 
                    where=(y1>=y2), 
                    color='gold', 
                    linewidth=0
                   )
ax[2].fill_betweenx(dfNewValues[:,0],
                    y1, 
                    y2, 
                    where=(y1<y2), 
                    color='lime', 
                    linewidth=0
                   )
    
plt.subplots_adjust(wspace=0)
plt.show()
```
